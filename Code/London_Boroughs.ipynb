**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
new1 = df[df['London_Boroughs'] == 'City of London']
n= new1.plot(kind='line', x='Month of Year', y='Average Price')
n.set_ylabel('Price')
```

```python
df['Year']= df['Month of Year'].apply(lambda x: x.year)
df
```

```python
df1 = df.groupby(by=['London_Boroughs', 'Year']).mean()
df1
```

```python
df1.sample(10)
```

```python
df1= df1.reset_index()
df1.head(10)
```

**Modeling**

creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

```python
def create_price_ratio(d):
    year_1998 = float(d['Average Price'][d['Year'] == 1998])
    year_2018 = float(d['Average Price'][d['Year'] == 2018])
    ratio = [year_2018 / year_1998]
    return ratio
```

```python
create_price_ratio(df1[df1['London_Boroughs']=='Camden'])
```

```python
create_price_ratio(df1[df1['London_Boroughs']=='Waltham Forest'])
```

```python
result = {}
```

```python
for i in df1['London_Boroughs'].unique():
    borough = df1[df1['London_Boroughs'] == i]
    result[i]= create_price_ratio(borough)
```

```python
result
```

```python
df_output = pd.DataFrame(result)
df_output
```

```python
df_output_T = df_output.T
df_output = df_output_T.reset_index()
```

```python
df_output
```

```python
df_output.head()
```

```python
df_output.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
df_output.head()
```

```python
top_boroughs= df_output.sort_values(by='2018', ascending=False)
top_boroughs
```

# CONCLUSION:

After analysing the all 33 London Boroughs we came to a conclusion that over the last 2 decades from 1998 to 2018 the greatest increase in the housing prices, on average were in the boroughs: 'Hackney','Waltham Forest','Southwark', 'Lewisham'.
The boroughs with the least increase in housing prices, on average, were: 'Hounslow', 'Richmond upon Thames', 'Harrow', 'Bromley'. Thus these 5 boroughs are best areas to live in due to rising rent prices.
